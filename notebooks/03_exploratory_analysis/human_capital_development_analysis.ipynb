```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import warnings
import os

warnings.filterwarnings("ignore")

# Set style for matplotlib
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

# Set display options
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)

print("‚úÖ Libraries imported successfully")
print("\n" + "="*70)
print("üìä EXPLORATORY DATA ANALYSIS (EDA)")
print("="*70)
```

```python
print("\nüì• Loading cleaned datasets...")

dataset2_cleaned_path = "../../data/cleaned/resources_human_capital_cleaned.csv"

dataset2_std_path = "../../data/cleaned/resources_human_capital_standardized.csv"

df_resources = pd.read_csv(dataset2_cleaned_path)

df_resources_std = pd.read_csv(dataset2_std_path)

print(f"\n‚úÖ Dataset loaded successfully!")
print(f"\n   Dataset 2: Resources & Human Capital")
print(f"      Shape: {df_resources.shape}")
print(f"      Countries: {df_resources['Country'].nunique()}")

# Create output directory for plots
output_dir = "../../outputs/eda"
os.makedirs(output_dir, exist_ok=True)
print(f"\n   üìÅ Output directory: {output_dir}")
```

```python
print("\n" + "="*70)
print("üìä BASIC STATISTICAL SUMMARY")
print("="*70)

# Dataset 2 - Resources & Human Capital
print("\n" + "-" * 70)
print("\nüìà DATASET 2: Resources & Human Capital")
print("-" * 70)

numeric_cols_d2 = df_resources.select_dtypes(include=[np.number]).columns

print(f"\nNumber of numeric features: {len(numeric_cols_d2)}")
print(f"\nDescriptive Statistics (first 10 features):")
display(df_resources[numeric_cols_d2[:10]].describe().T)

print(f"\nüìä Data Quality Check:")
print(f"   Total rows: {len(df_resources)}")
print(f"   Total columns: {len(df_resources.columns)}")
print(f"   Missing values: {df_resources.isnull().sum().sum()}")
```

```python
print("\n" + "="*70)
print("üìä DISTRIBUTION ANALYSIS - DATASET 2")
print("="*70)

# Select key resource indicators
key_indicators_d2 = [
    col for col in numeric_cols_d2 
    if any(keyword in col for keyword in ['Population', 'Labor', 'Energy', 'Edu'])
][:12]

if len(key_indicators_d2) > 0:
    print(f"\nüìä Creating histograms for {len(key_indicators_d2)} key indicators...")
    
    n_cols = 4
    n_rows = (len(key_indicators_d2) + n_cols - 1) // n_cols
    
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, n_rows * 4))
    axes = axes.flatten() if n_rows > 1 else [axes] if n_cols == 1 else axes
    
    for idx, col in enumerate(key_indicators_d2):
        ax = axes[idx]
        
        data = df_resources[col].dropna()
        ax.hist(data, bins=30, color='coral', edgecolor='black', alpha=0.7)
        
        mean_val = data.mean()
        median_val = data.median()
        
        ax.axvline(mean_val, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_val:.2f}')
        ax.axvline(median_val, color='green', linestyle='--', linewidth=2, label=f'Median: {median_val:.2f}')
        
        ax.set_title(col[:50] + ('...' if len(col) > 50 else ''), fontsize=9, fontweight='bold')
        ax.set_xlabel('Value', fontsize=8)
        ax.set_ylabel('Frequency', fontsize=8)
        ax.legend(fontsize=7)
        ax.grid(axis='y', alpha=0.3)
    
    for idx in range(len(key_indicators_d2), len(axes)):
        axes[idx].set_visible(False)
    
    plt.tight_layout()
    plt.savefig(f'{output_dir}/resources_human/distributions_resources_indicators.png', dpi=300, bbox_inches='tight')
    print(f"   ‚úÖ Saved: {output_dir}/distributions_resources_indicators.png")
    plt.show()
else:
    print("   ‚ö†Ô∏è No suitable columns found for visualization")
```

```python
print("\n" + "="*70)
print("üî• CORRELATION ANALYSIS - DATASET 2")
print("="*70)

print("\nüìä Creating correlation heatmap for Resources & Human Capital...")

# Select important features
important_cols_d2 = [col for col in numeric_cols_d2 if any(
    keyword in col for keyword in ['Population', 'Labor', 'Energy', 'Edu', 'Rural', 'Agri']
)][:20]

if len(important_cols_d2) > 0:
    corr_matrix_d2 = df_resources[important_cols_d2].corr()
    
    fig, ax = plt.subplots(figsize=(16, 14))
    mask = np.triu(np.ones_like(corr_matrix_d2, dtype=bool))
    
    sns.heatmap(corr_matrix_d2, 
                mask=mask,
                annot=True, 
                fmt='.2f', 
                cmap='coolwarm', 
                center=0,
                square=True,
                linewidths=0.5,
                cbar_kws={"shrink": 0.8},
                annot_kws={'size': 7},
                ax=ax)
    
    ax.set_title('Correlation Heatmap - Resources & Human Capital (Top 20 Features)', 
                 fontsize=14, fontweight='bold', pad=20)
    
    plt.xticks(rotation=45, ha='right', fontsize=8)
    plt.yticks(rotation=0, fontsize=8)
    
    plt.tight_layout()
    plt.savefig(f'{output_dir}/resources_human/correlation_heatmap_resources.png', dpi=300, bbox_inches='tight')
    print(f"   ‚úÖ Saved: {output_dir}/correlation_heatmap_resources.png")
    plt.show()
    
    # Find strongest correlations
    print("\nüìä Strongest Positive Correlations (> 0.7):")
    strong_corr_d2 = []
    for i in range(len(corr_matrix_d2.columns)):
        for j in range(i+1, len(corr_matrix_d2.columns)):
            if abs(corr_matrix_d2.iloc[i, j]) > 0.7:
                strong_corr_d2.append({
                    'Feature 1': corr_matrix_d2.columns[i],
                    'Feature 2': corr_matrix_d2.columns[j],
                    'Correlation': corr_matrix_d2.iloc[i, j]
                })
    
    if strong_corr_d2:
        strong_corr_df2 = pd.DataFrame(strong_corr_d2).sort_values('Correlation', ascending=False)
        display(strong_corr_df2.head(10))
    else:
        print("   No correlations > 0.7 found")
else:
    print("   ‚ö†Ô∏è No suitable columns found for correlation analysis")
```

```python
print("\n" + "="*70)
print("üì¶ BOX PLOT ANALYSIS - DATASET 2")
print("="*70)

print("\nüìä Creating box plots for key resource indicators...")

box_cols_d2 = [col for col in numeric_cols_d2 if any(
    keyword in col for keyword in ['Population', 'Labor', 'Energy', 'Edu']
)][:8]

if len(box_cols_d2) > 0:
    n_cols = 4
    n_rows = (len(box_cols_d2) + n_cols - 1) // n_cols
    
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, n_rows * 4))
    axes = axes.flatten() if n_rows > 1 else [axes] if n_cols == 1 else axes
    
    for idx, col in enumerate(box_cols_d2):
        ax = axes[idx]
        
        data = df_resources[col].dropna()
        
        bp = ax.boxplot([data], 
                        labels=[''],
                        patch_artist=True,
                        boxprops=dict(facecolor='coral', alpha=0.7),
                        medianprops=dict(color='red', linewidth=2),
                        whiskerprops=dict(linewidth=1.5),
                        capprops=dict(linewidth=1.5))
        
        stats_text = f"Mean: {data.mean():.2f}\nMedian: {data.median():.2f}\nStd: {data.std():.2f}"
        ax.text(0.98, 0.98, stats_text,
                transform=ax.transAxes,
                verticalalignment='top',
                horizontalalignment='right',
                bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5),
                fontsize=8)
        
        ax.set_title(col[:50] + ('...' if len(col) > 50 else ''), 
                    fontsize=9, fontweight='bold')
        ax.set_ylabel('Value', fontsize=8)
        ax.grid(axis='y', alpha=0.3)
    
    for idx in range(len(box_cols_d2), len(axes)):
        axes[idx].set_visible(False)
    
    plt.tight_layout()
    plt.savefig(f'{output_dir}/resources_human/boxplots_resources_indicators.png', dpi=300, bbox_inches='tight')
    print(f"   ‚úÖ Saved: {output_dir}/boxplots_resources_indicators.png")
    plt.show()
else:
    print("   ‚ö†Ô∏è No suitable columns found for box plots")
```

```python
print("\n" + "="*70)
print("üîÄ SCATTER PLOT MATRIX - DATASET 2")
print("="*70)

print("\nüìä Creating pair plot for top resource indicators...")

# Select top 5 key resource indicators
pair_cols_d2 = [col for col in numeric_cols_d2 if any(
    keyword.lower() in col.lower() for keyword in ['population', 'labor', 'energy']
)][:5]

if len(pair_cols_d2) >= 2:
    print(f"   Analyzing {len(pair_cols_d2)} variables...")
    
    # Copy selected columns
    df_pair_d2 = df_resources[pair_cols_d2].copy()
    
    # Ensure all columns are numeric
    df_pair_d2 = df_pair_d2.apply(pd.to_numeric, errors='coerce')
    
    # Drop columns that are all NaN
    df_pair_d2 = df_pair_d2.dropna(axis=1, how='all')
    
    # Check if at least 2 columns remain
    if df_pair_d2.shape[1] >= 2:
        # Make column names short and unique
        df_pair_d2.columns = [f"Var_{i+1}" for i in range(df_pair_d2.shape[1])]
        
        # Create the pair plot
        pairplot = sns.pairplot(df_pair_d2,
                                diag_kind='kde',
                                plot_kws={'alpha': 0.6, 's': 30, 'edgecolor': 'k', 'color': 'coral'},
                                diag_kws={'shade': True, 'color': 'coral'})
        
        pairplot.fig.suptitle('Pair Plot - Resources & Human Capital', 
                              fontsize=14, fontweight='bold', y=1.02)
        
        plt.tight_layout()
        save_path = f'{output_dir}/resources_human/pairplot_resources_indicators.png'
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"   ‚úÖ Saved: {save_path}")
        plt.show()
    else:
        print("   ‚ö†Ô∏è Not enough valid numeric columns after cleaning")
else:
    print("   ‚ö†Ô∏è Not enough columns for pair plot")
```

```python
print("\n" + "="*70)
print("üìä TOP & BOTTOM COUNTRIES COMPARISON - DATASET 2")
print("="*70)

# Select a key indicator for Dataset 2 (e.g., first numeric column related to Population, Labor, or Energy)
indicator_cols_d2 = [col for col in numeric_cols_d2 if any(
    keyword.lower() in col.lower() for keyword in ['population', 'labor', 'energy']
)]

if indicator_cols_d2:
    indicator_col = indicator_cols_d2[0]
    print(f"\nüìä Analyzing: {indicator_col}")
    
    # Ensure numeric
    df_resources[indicator_col] = pd.to_numeric(df_resources[indicator_col], errors='coerce')
    
    # Get top 15 and bottom 15 countries
    df_sorted = df_resources[['Country', indicator_col]].dropna().sort_values(indicator_col, ascending=False)
    
    top_15 = df_sorted.head(15)
    bottom_15 = df_sorted.tail(15)
    
    fig, axes = plt.subplots(1, 2, figsize=(22, 10))
    
    # Top 15 countries
    ax1 = axes[0]
    bars1 = ax1.barh(range(len(top_15)), top_15[indicator_col], color='green', alpha=0.7)
    ax1.set_yticks(range(len(top_15)))
    ax1.set_yticklabels(top_15['Country'], fontsize=9)
    ax1.set_xlabel(indicator_col[:50], fontsize=10)
    ax1.set_title('Top 15 Countries', fontsize=12, fontweight='bold')
    ax1.grid(axis='x', alpha=0.3)
    ax1.invert_yaxis()
    
    for i, value in enumerate(top_15[indicator_col]):
        ax1.text(value, i, f' {value:,.1f}', va='center', fontsize=8)
    
    # Bottom 15 countries
    ax2 = axes[1]
    bars2 = ax2.barh(range(len(bottom_15)), bottom_15[indicator_col], color='red', alpha=0.7)
    ax2.set_yticks(range(len(bottom_15)))
    ax2.set_yticklabels(bottom_15['Country'], fontsize=9)
    ax2.set_xlabel(indicator_col[:50], fontsize=10)
    ax2.set_title('Bottom 15 Countries', fontsize=12, fontweight='bold')
    ax2.grid(axis='x', alpha=0.3)
    ax2.invert_yaxis()
    
    for i, value in enumerate(bottom_15[indicator_col]):
        ax2.text(value, i, f' {value:,.1f}', va='center', fontsize=8)
    
    plt.tight_layout()
    plt.suptitle(f'Top & Bottom 15 Countries - {indicator_col}', fontsize=16, fontweight='bold', y=1.02)
    
    # Save the figure
    save_path = f'{output_dir}/resources_human/top_bottom_15_{indicator_col.replace(" ", "_")}.png'
    plt.savefig(save_path, dpi=300, bbox_inches='tight')
    print(f"‚úÖ Saved figure: {save_path}")
    
    plt.show()
    
else:
    print("‚ö†Ô∏è No suitable indicator column found in Dataset 2")
```
